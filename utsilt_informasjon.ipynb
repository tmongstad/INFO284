```python
# Hvor mange fisketokter som er registrert for hvilken type fisk. (de 1000 minste er silt ut)
unique_fish_types = df.groupby(['Art - FDIR', 'Art - FDIR (kode)'], ).size()
unique_fish_types_df = unique_fish_types.reset_index(name='Count')  # R
unique_fish_types_sorted = unique_fish_types_df.sort_values(by = 'Count', ascending = False, )
filtered_df = unique_fish_types_sorted[unique_fish_types_sorted['Count'] <= 100]
filtered_df
```

```python
fish_weigh = df.groupby(['Art - FDIR', 'Art - FDIR (kode)'], ).size()
fish_weigh_df = fish_weigh.reset_index(name='Count')  # R
fish_weigh_sorted = fish_weigh_df.sort_values(by = 'Count', ascending = False, )
filtered_df = fish_weigh_sorted[fish_weigh_sorted['Count'] >= 1000]
filtered_df
```

```python
average_weight_per_fish_type = df.groupby(['Art - FDIR', 'Art - FDIR (kode)'])['Rundvekt'].mean().reset_index(name='Average Weight')
average_weight_per_fish_type.sort_values
```

```python
# Slå sammen Bruttotonnasjekolonnene - Tar utgangspunkt i at det ikke er særlig stor forskjell på Bruttotonnasje 1969 og 
# Bruttotonnasje annen, da dette bare er forskjellige klassifiseringsmetoder for lastekapasitet(Hulrom i skipet).

df['Bruttotonnasje Kombinert'] = df['Bruttotonnasje 1969'].fillna(df['Bruttotonnasje annen'])

# Dropper unødvendig informasjon. Tabellen inneholder masse duplikater. som egen kolonne for navn og for kode
# Forskjellige klassifiseringsmåter. Holder på FDIR sine.

df.drop(columns = ['Meldingsdato', 
                   'Meldingsklokkeslett', 
                   'Startdato', 
                   'Startklokkeslett', 
                   'Hovedområde start', 
                   'Lokasjon start (kode)', 
                   'Stoppdato', 
                   'Stoppklokkeslett', 
                   'Fangstår', 
                   'Hovedområde stopp (kode)', 
                   'Hovedområde stopp', 
                   'Lokasjon stopp (kode)', 
                   'Redskap FAO (kode)', 
                   'Redskap FAO', 
                   'Redskap FDIR', 
                   'Hovedart FAO (kode)', 
                   'Hovedart FAO', 
                   'Art FAO (kode)',
                   'Art FAO',
                   'Art - FDIR',
                   'Art - gruppe', 
                   'Lengdegruppe', 
                   'Bredde', 
                   'Fartøylengde',
                   'Bruttotonnasje 1969',
                   'Bruttotonnasje annen'
    ], inplace=True)
```

```python
filtered_df = df[(df['Redskap FDIR (kode)'] == 51)]

def join_unique_values(series):
    return ', '.join(series.dropna().astype(str).unique())

# Aggregating data
aggregated_df = filtered_df.groupby(['Melding ID', 'Starttidspunkt'], as_index=False).agg({
    'Startposisjon bredde': 'first',  # Antar at verdiene er lik for dei ulike radene. Tar med informasjon frå første raden som slås sammen
    'Startposisjon lengde': 'first',
    'Havdybde start': 'first',
    'Stopptidspunkt': 'first',
    'Varighet': 'last', 
    'Stopposisjon bredde': 'last',  # Slår sammen og bruker verdien fra siste raden.
    'Stopposisjon lengde': 'last',
    'Havdybde stopp': 'last',
    'Trekkavstand': 'last',  # summer opp dei ulike radene.
    'Redskap FDIR (kode)': 'last',
    'Hovedart - FDIR (kode)': join_unique_values, # Tar med verdiene fra dei ulike radene og slår dei sammen til ein tekststreng.
    'Art - FDIR (kode)': join_unique_values,
    'Art - gruppe (kode)': join_unique_values,
    'Rundvekt': 'sum',  # Summing up the total weight
    'Lengdegruppe (kode)': 'first',  # Assuming to take the first value
    'Bruttotonnasje Kombinert': 'first'  # Assuming to take the first value
})

aggregated_df
```

```python
aggregated_df = aggregated_df.groupby(['Melding ID'], as_index=False).agg({
    'Startposisjon bredde': 'first',  # Antar at verdiene er lik for dei ulike radene. Tar med informasjon frå første raden som slås sammen
    'Startposisjon lengde': 'first',
    'Starttidspunkt': 'first',
    'Havdybde start': 'first',
    'Stopptidspunkt': 'last',
    'Varighet': 'sum', 
    'Stopposisjon bredde': 'last',  # Slår sammen og bruker verdien fra siste raden.
    'Stopposisjon lengde': 'last',
    'Havdybde stopp': 'last',
    'Trekkavstand': 'sum',  # summer opp dei ulike radene.
    'Redskap FDIR (kode)': join_unique_values,
    'Hovedart - FDIR (kode)': join_unique_values, # Tar med verdiene fra dei ulike radene og slår dei sammen til ein tekststreng.
    'Art - FDIR (kode)': join_unique_values,
    'Art - gruppe (kode)': join_unique_values,
    'Rundvekt': 'sum',  # Summing up the total weight
    'Lengdegruppe (kode)': 'first',  # Assuming to take the first value
    'Bruttotonnasje Kombinert': 'first'  # Assuming to take the first value
})

aggregated_df
```

```python
import numpy as np

# Define your bins. The last bin will capture everything over 1500.
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1300, 1400, 1500, np.inf]
# Define your labels. These should be one fewer than the bins since they label the intervals between the bin edges.
labels = range(1, len(bins))  # This will create numerical labels 1 through 16

# Categorize 'Bruttotonnasje Kombinert' into the defined bins
aggregated_df['Bruttotonnasje Group'] = pd.cut(aggregated_df['Bruttotonnasje Kombinert'], bins=bins, labels=labels)

# Check the first few rows to verify the new colum
aggregated_df.head()
```
